```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.colors as mcolors
```

```python
num_cur = 166
num_frm = 108
num_prj = 108
num_ppr = 200

th = '70'

is_keyword_matching = False

if (is_keyword_matching):
    KM = "KM_"
else:
    KM = ""
```

```python
cur_df = pd.read_excel(th +'/SDG_curricula_' + KM + th +'.xlsx')
frm_df = pd.read_excel(th +'/SDG_frameworks_' + KM + th +'.xlsx')
prj_df = pd.read_excel('data/SDG_Project_meta.xlsx')
ppr_df = pd.read_excel(th +'/SDG_papers_' + KM + th +'.xlsx')
```

```python
columns = ['sdg1', 'sdg2', 'sdg3', 'sdg4', 'sdg5', 'sdg6', 'sdg7', 'sdg8', 'sdg9', 'sdg10', 'sdg11', 'sdg12', 'sdg13', 'sdg14', 'sdg15', 'sdg16', 'sdg17', ]
```
